#### Questão 04

\begin{equation*}
\begin{aligned}
\text{Variáveis de decisão:} \\
\text{x1:} & & \text{quantidade de compra de caminhões grandes}  \\
\text{x2:} & & \text{quantidade de  compra de caminhões pequenos}  \\\\
\text{Maximizar} & & Z=5 \cdot X_1 + 4 \cdot X_2\\\\
\text{Restrições:} \\
\text{Orçamento:} & & 50000 \cdot X_1 + 30000 \cdot X_2 + S_0 = 200000 \\
\text{Toneladas de mercadoria:} & & 10 \cdot X_1 + 5 \cdot X_2 + S_1 = 30 \\
\text{Não-negatividade:} & & X_1, X_2, S_0, S_1 \geq 0 \\
\end{aligned}
\end{equation*}

```python
from cvxopt.modeling import op, variable
from plotly import graph_objects as go
```

```python
cam_grande = variable(1, 'quantidade de compra de caminhões grandes')
cam_peq = variable(1, 'quantidade de  compra de caminhões pequenos')
folga = variable(2, 'Variável de folga')
```

```python
fob = - 5 * cam_grande[0] - 4 * cam_peq[0]
```

```python
restricoes = []

restricoes.append(50000 * cam_grande[0] + 30000 * cam_peq[0] + folga[0] == 200000)
restricoes.append(10 * cam_grande[0] + 5 * cam_peq[0] + folga[1] == 30)
restricoes.append(cam_grande[0] >= 0)
restricoes.append(cam_peq[0] >= 0)
restricoes.append(folga[0] >= 0)
restricoes.append(folga[1] >= 0)
```

```python
empresa_transporte = op(fob, restricoes)
empresa_transporte.solve('dense', 'glpk')
```

```python
print(f'Lucro projetado para empresa de transporte é de ${-fob.value()[0]:.2f}')
print(f'{cam_grande.name} é de {cam_grande[0].value()[0]}')
print(f'{cam_peq.name} é de {cam_peq[0].value()[0]}')
```

```python
print('Variável dual associado ao limite de orçamento', restricoes[0].multiplier.value)
print('Variável dual associado ao limite de toneladas de mercadoria', restricoes[1].multiplier.value)
print('Variável dual associado ao limite de não-negatividade caminhões grandes (x1)', restricoes[2].multiplier.value)
print('Variável dual associado ao limite de não-negatividade caminhões pequenos (x2)', restricoes[2].multiplier.value)
```

```python
figura = go.Figure(data=go.Bar(y=[cam_grande[0].value()[0], cam_peq[0].value()[0]], x=['caminhao_grande','caminhao_pequeno']))
figura.show()
```

```python
from cvxopt import matrix, solvers
import numpy as np
```

```python
c = matrix([-5.0, -4.0, 0.0, 0.0])

Aeq = matrix([
    [50000.0, 30000.0, 1.0, 0.0],
    [10.0, 5.0, 0.0, 1.0]], (2,4), 'd')

beq = matrix([200000.0, 30.0])

G = matrix(-np.eye(4))
h = matrix(np.zeros(4))

print(G)

sol = solvers.lp(c, G, h, Aeq, beq)
print(sol['x'])
```
