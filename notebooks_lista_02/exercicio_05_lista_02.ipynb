#### Questão 05

\begin{equation*}
\begin{aligned}
\text{Variáveis de decisão:} \\
\text{x1:} & & \text{quantidade de vendas produto A}  \\
\text{x2:} & & \text{quantidade de vendas produto B}  \\\\
\text{Maximizar} & & Z=3 \cdot X_1 + 5 \cdot X_2\\\\
\text{Restrições:} \\
\text{m² de espaço de prateleira:} & & 2 \cdot X_1 + 3 \cdot X_2 + S_0 = 60 \\
\text{Não-negatividade:} & & X_1, X_2, S_0 \geq 0 \\
\end{aligned}
\end{equation*}

```python
from cvxopt.modeling import op, variable
from plotly import graph_objects as go
```

```python
prod_a = variable(1, 'quantidade de vendas produto A')
prod_b = variable(1, 'quantidade de vendas produto B')
folga = variable(1, 'Variável de folga')
```

```python
fob = - 3 * prod_a[0] - 5 * prod_b[0]
```

```python
restricoes = []

restricoes.append(2 * prod_a[0] + 3 * prod_b[0] + folga[0] == 60)
restricoes.append(prod_a[0] >= 0)
restricoes.append(prod_b[0] >= 0)
restricoes.append(folga[0] >= 0)
```

```python
loja = op(fob, restricoes)
loja.solve('dense', 'glpk')
```

```python
print(f'Lucro projetado para loja é de ${-fob.value()[0]:.2f}')
print(f'{prod_a.name} é de {prod_a[0].value()[0]}')
print(f'{prod_b.name} é de {prod_b[0].value()[0]}')
```

```python
print('Variável dual associado ao limite de m² de espaço de prateleira', restricoes[0].multiplier.value)
print('Variável dual associado ao limite de não-negatividade produto A (x1)', restricoes[2].multiplier.value)
print('Variável dual associado ao limite de não-negatividade produto B (x2)', restricoes[2].multiplier.value)
```

```python
figura = go.Figure(data=go.Bar(y=[prod_a[0].value()[0], prod_b[0].value()[0]], x=['produto_a','produto_b']))
figura.show()
```

```python
from cvxopt import matrix, solvers
import numpy as np
```

```python
c = matrix([-3.0, -5.0, 0.0])

Aeq = matrix([[2.0, 3.0, 1.0]], (1,3), 'd')

beq = matrix([60.0])

G = matrix(-np.eye(3))
h = matrix(np.zeros(3))

print(G)

sol = solvers.lp(c, G, h, Aeq, beq)
print(sol['x'])
```
