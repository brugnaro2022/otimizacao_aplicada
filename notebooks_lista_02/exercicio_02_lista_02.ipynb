#### Questão 02

\begin{equation*}
\begin{aligned}
\text{Variáveis de decisão:} \\
\text{x1:} & & \text{quantidade de produção brinquedo A}  \\
\text{x2:} & & \text{quantidade de produção brinquedo B}  \\\\
\text{Maximizar} & & Z=10 \cdot X_1 + 15 \cdot X_2\\\\
\text{Restrições:} \\
\text{Horas de trabalho:} & & 2 \cdot X_1 + X_2 + S_0 = 80 \\
\text{Horas de máquina:} & & X_1 + 3 \cdot X_2 + S_1 = 40 \\
\text{Não-negatividade:} & & X_1, X_2, S_0, S_1 \geq 0 \\
\end{aligned}
\end{equation*}

```python
from cvxopt.modeling import op, variable
from plotly import graph_objects as go
```

```python
brinq_a = variable(1, 'quantidade de produção brinquedo A')
brinq_b = variable(1, 'quantidade de produção brinquedo B')
folga = variable(2, 'Variável de folga')
```

```python
fob = - 10 * brinq_a[0] - 15 * brinq_b[0]
```

```python
restricoes = []
restricoes.append(2 * brinq_a[0] + brinq_b[0] + folga[0] == 80)
restricoes.append(brinq_a[0] + 3 * brinq_b[0] + folga[1] == 40)
restricoes.append(brinq_a[0] >= 0)
restricoes.append(brinq_b[0] >= 0)
restricoes.append(folga[0] >= 0)
restricoes.append(folga[1] >= 0)
```

```python
fabrica = op(fob, restricoes)
fabrica.solve('dense', 'glpk')
```

```python
print(f'Lucro projetado para a fábrica de brinquedos é de ${-fob.value()[0]:.2f}')
print(f'{brinq_a.name} é de {brinq_a[0].value()[0]}')
print(f'{brinq_b.name} é de {brinq_b[0].value()[0]}')
```

```python
print('Variável dual associado ao limite de horas de trabalho', restricoes[0].multiplier.value)
print('Variável dual associado ao limite de horas de máquina', restricoes[1].multiplier.value)
print('Variável dual associado ao limite de não-negatividade brinquedo A (x1)', restricoes[2].multiplier.value)
print('Variável dual associado ao limite de não-negatividade brinquedo B (x2)', restricoes[2].multiplier.value)
```

```python
figura = go.Figure(data=go.Bar(y=[brinq_a[0].value()[0], brinq_b[0].value()[0]], x=['brinquedo_a','brinquedo_b']))
figura.show()
```
