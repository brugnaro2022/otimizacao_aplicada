#### Questão 03

\begin{equation*}
\begin{aligned}
\text{Variáveis de decisão:} \\
\text{x1:} & & \text{quantidade de mesas produzidas}  \\
\text{x2:} & & \text{quantidade de cadeiras produzidas}  \\\\
\text{Maximizar} & & Z=200 \cdot X_1 + 100 \cdot X_2\\\\
\text{Restrições:} \\
\text{Metros de madeira:} & & 4 \cdot X_1 + 2 \cdot X_2 + S_0 = 800 \\
\text{Horas de mão de obra:} & & 2 \cdot X_1 + X_2 + S_1 = 400 \\
\text{Não-negatividade:} & & X_1, X_2, S_0, S_1 \geq 0 \\
\end{aligned}
\end{equation*}

```python
from cvxopt.modeling import op, variable
from plotly import graph_objects as go
```

```python
mesa = variable(1, 'quantidade de mesas produzidas')
cadeira = variable(1, 'quantidade de cadeiras produzidas')
folga = variable(2, 'Variável de folga')
```

```python
fob = - 200 * mesa[0] - 100 * cadeira[0]
```

```python
restricoes = []

restricoes.append(4 * mesa[0] + 2 * cadeira[0] + folga[0] == 800)
restricoes.append(2 * mesa[0] + cadeira[0] + folga[1] == 400)
restricoes.append(mesa[0] >= 0)
restricoes.append(cadeira[0] >= 0)
restricoes.append(folga[0] >= 0)
restricoes.append(folga[1] >= 0)
```

```python
empresa = op(fob, restricoes)
empresa.solve('dense', 'glpk')
```

```python
print(f'Lucro projetado para empresa é de ${-fob.value()[0]:.2f}')
print(f'{mesa.name} é de {mesa[0].value()[0]}')
print(f'{cadeira.name} é de {cadeira[0].value()[0]}')
```

```python
print('Variável dual associado ao limite de metros de madeira', restricoes[0].multiplier.value)
print('Variável dual associado ao limite de horas de mão de obra', restricoes[1].multiplier.value)
print('Variável dual associado ao limite de não-negatividade mesa (x1)', restricoes[2].multiplier.value)
print('Variável dual associado ao limite de não-negatividade cadeira (x2)', restricoes[2].multiplier.value)
```

```python
figura = go.Figure(data=go.Bar(y=[mesa[0].value()[0], cadeira[0].value()[0]], x=['mesa','cadeira']))
figura.show()
```

```python
from cvxopt import matrix, solvers
import numpy as np
```

```python
c = matrix([-200.0, -100.0, 0.0, 0.0])

Aeq = matrix([
    [4.0, 2.0, 1.0, 0.0],
    [2.0, 1.0, 0.0, 1.0]], (2,4), 'd')

beq = matrix([800.0, 400.0])

G = matrix(-np.eye(4))
h = matrix(np.zeros(4))

print(G)

sol = solvers.lp(c, G, h, Aeq, beq)
print(sol['x'])
```
