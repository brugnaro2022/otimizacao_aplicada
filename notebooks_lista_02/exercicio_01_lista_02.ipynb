#### Questão 01

\begin{equation*}
\begin{aligned}
\text{Variáveis de decisão:} \\
\text{x1:} & & \text{quantidade de acres de milho plantar}  \\
\text{x2:} & & \text{quantidade de acres de trigo plantar}  \\\\
\text{Maximizar} & & Z=70 \cdot X_1 + 50 \cdot X_2\\\\
\text{Restrições:} \\
\text{Horas de trabalho:} & & 2 \cdot X_1 + X_2 + S_0 = 800 \\
\text{Orçamento:} & & 40 \cdot X_1 + 30 \cdot X_2 + S_1 = 2.400 \\
\text{Não-negatividade:} & & X_1, X_2, S_0, S_1 \geq 0 \\
\end{aligned}
\end{equation*}

```python
from cvxopt.modeling import op, variable
from plotly import graph_objects as go
```

```python
milho = variable(1, 'Quantidade de acres de milho plantar')
trigo = variable(1, 'Quantidade de acres de trigo plantar')
folga = variable(2, 'Variável de folga')
```

```python
fob = - 70 * milho[0] - 50 * trigo[0]
```

```python
restricoes = []
restricoes.append(2 * milho[0] + trigo[0] + folga[0] == 800)
restricoes.append(40 * milho[0] + 30 * trigo[0] + folga[1] == 2400)
restricoes.append(milho[0] >= 0)
restricoes.append(trigo[0] >= 0)
restricoes.append(folga[0] >= 0)
restricoes.append(folga[1] >= 0)
```

```python
fazendeiro = op(fob, restricoes)
fazendeiro.solve('dense', 'glpk')
```

```python
print(f'Lucro projetado para o fazendeiro é de ${-fob.value()[0]:.2f}')
print(f'{milho.name} é de {milho[0].value()[0]}')
print(f'{trigo.name} é de {trigo[0].value()[0]}')
```

```python
print('Variável dual associado ao limite de horas de trabalho', restricoes[0].multiplier.value)
print('Variável dual associado ao limite de orçamento', restricoes[1].multiplier.value)
print('Variável dual associado ao limite de não-negatividade milho (x1)', restricoes[2].multiplier.value)
print('Variável dual associado ao limite de não-negatividade trigo (x2)', restricoes[2].multiplier.value)
```

```python
figura = go.Figure(data=go.Bar(y=[milho[0].value()[0], trigo[0].value()[0]], x=['Milho','Trigo']))
figura.show()
```

```python
from cvxopt import matrix, solvers
import numpy as np
```

```python
c = matrix([-70.0, -50.0, 0.0, 0.0])
Aeq = matrix([[2.0, 1.0, 1.0, 0.0], [40.0, 30.0, 0.0, 1.0]], (2,4), 'd')
beq = matrix([800.0, 2400.0])

G = matrix(-np.eye(4))
h = matrix(np.zeros(4))

print(G)

sol = solvers.lp(c, G, h, Aeq, beq)
print(sol['x'])
```
